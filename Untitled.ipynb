```python
import os
import numpy as np
from skimage.exposure import rescale_intensity
from skimage.io import imread, imsave
from skimage.transform import resize, rescale, rotate
import pydicom
from matplotlib import pyplot as plt

# volume preprocessing methods
def crop_sample(image):
    # cleaning
    image[image < np.max(image) * 0.1] = 0
    # crop 
    y_projection = np.max(image, axis=0)
    y_nonzero = np.nonzero(y_projection)
    y_min = np.min(y_nonzero)
    y_max = np.max(y_nonzero) + 1
    x_projection = np.max(image, axis=1)
    x_nonzero = np.nonzero(x_projection)
    x_min = np.min(x_nonzero)
    x_max = np.max(x_nonzero) + 1
    return image[x_min:x_max, y_min:y_max]

# resizing the volume while preserving the z-axis
def resize_sample(image, size=224):
    v_shape = image.shape
    out_shape = (size, size)

    image = resize(
        image,
        output_shape=out_shape,
        order=2,
        mode="constant",
        cval=0,
        anti_aliasing=False,
    )
    return image

# z score normalization
def normalize_volume(image):
    p10 = np.percentile(image, 10)
    p99 = np.percentile(image, 99)
    image = rescale_intensity(image, in_range=(p10, p99))
    m = np.mean(image, axis=(0, 1))
    s = np.std(image, axis=(0, 1))
    image = (image - m) / s
    return image
```

```python
curr_dir = os.getcwd()
```

```python
dicom_path = curr_dir + '/5122684/' + 'ser700img00001.dcm'
tmp = pydicom.read_file(dicom_path)
print(np.shape(tmp.pixel_array))
```

```python
# original image
plt.imshow(tmp.pixel_array, cmap='gray')
plt.show()

# crop
crop_image = crop_sample(tmp.pixel_array)
plt.imshow(crop_image, cmap='gray')
plt.show()

# resize
resize_image = resize_sample(crop_image)
plt.imshow(resize_image, cmap='gray')
plt.show()

# normalize
norm_image = normalize_volume(crop_image)
plt.imshow(norm_image, cmap='gray')
plt.show()
plt.hist(norm_image.reshape(-1,),density=True,bins)
plt.show()
```
